Resising the image using pillow library

```python
from PIL import Image 
```

```python
img=Image.open('/content/angela.png')
img_resized=img.resize((200,200))
```

Converting RGB to greyscale images using Open Cv

```python
imgs=cv2.imread('/content/angela.png')
```

```python
type(imgs)
```
